```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys
```

```python
directory = "/Users/brent/target5and7data/testData2019/module3/T5TEATesting/20190813_1653/"

fig = plt.figure(figsize=(10,8))
pmtref = fig.add_subplot(111)
for i in range(0, 16, 4):
    data = np.loadtxt("{}PMTref4Scan_asic2_channel{}.txt".format(directory, i))
    asic = int(data[0][0])
    group = int(data[0][1]) / 4 + asic * 4
    dac_set = data[:, 2]
    meas = data[:, 3]
    la = "group {}".format(group)
    pmtref.plot(dac_set, meas, marker='o', label=la)
    pmtref.legend(loc=0)
    pmtref.set_xlabel("PMTref4 DAC Setting")
    pmtref.set_ylabel("Amon Output in V")
    pmtref.set_title("ASIC {}".format(asic))

plt.savefig("/Users/brent/Desktop/Grad School/Research/PMTref4Scan_asic{}.pdf".format(asic))
plt.show()
```
